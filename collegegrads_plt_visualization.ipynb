### Data Visualization - Scatter Plots

When plotting median salaries across sample sizes of majors, it seems popular majors don't seem to return higher paying full-time salaries. Conversely, lesser popular, smaller, highly-specialized majors have a higher occurance of higher median salaries.

```python
recent_grads.plot(x="Sample_size", y="Median", kind="scatter", title="Medians of Majors by Sample Size")
plt.show()
```

When plotting unemployment rate against sample size of majors, there appears to be a standard bell curve with a peak around 0.05%.

```python
recent_grads.plot(x="Sample_size", y="Unemployment_rate", kind="scatter", title="Unemployment Rate of Majors by Sample Size")
plt.show()
```

When plotting full-time employment againsts median salaries, it seems that higher salaries are more likely to be earned in non-full time roles (perhaps majors with high-paying consultants).

```python
recent_grads.plot(x="Full_time", y="Median", kind="scatter", title="Medians of Majors by Full-Time Employment Total")
plt.show()
```

Little to no trends appear when plotting unemployment rates against the share of women in majors.

```python
recent_grads.plot(x="ShareWomen", y="Unemployment_rate", kind="scatter", title="Unemployment Rate of Majors by Share of Women")
plt.show()
```

When compared against each other (gender against median salary), the following two graphs appear to support that men are paid higher than women in some majors.

```python
recent_grads.plot(x="Men", y="Median", kind="scatter", title="Median of Majors by Number of Men")
plt.show()
```

```python
recent_grads.plot(x="Women", y="Median", kind="scatter", title="Median of Majors by Number of Women")
plt.show()
```

### Data Visualization - Histograms`

Most graduates appear to have 0-500 people in the major.

```python
recent_grads["Sample_size"].hist(figsize=(10,5), range=(0,4500), bins=10)
```

Ranges nearing $40,000 is the most popular median salary in the dataset.

```python
recent_grads["Median"].hist(figsize=(10,5), range=(0,110000), bins=10)
```
